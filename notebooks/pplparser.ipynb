# Walkthrough on how to operate pplparser on jupyter notebook.
You first load the necessary libraries (sys and pandas), Then attach the path where we can find the folder that has our parsing files. After that import pplparser and ppbBatchParser classes from the pplparser file.

The next step is to add three path files. The first path is for a ppl file you want to analyze, second is for a csv file that has a table with varname,branchname,out_unit,out_unit_profile,time_unit of the variables you want to check out, and the third is for a csv file that has branch_in,branch_out for the branches you wish to join their variables.

After all that you have to load the csv files as dataframes using pandas and create an instance for the loaded class with the ppl file path as input.

```python
import sys
import pandas as pd
sys.path.append(r'C:\Users\HP\Desktop\Pontem Analytics\PyPontem\src\pypontem')
from ppl.pplparser import pplParser,pplBatchParser
filepath1 = r"C:\Users\HP\Desktop\Pontem Analytics\ppl\PPLfiles_batchparsing\GV-SC_Apr-25-SI_2.ppl"
pplparser = pplParser(filepath1)
input_matrix = pd.read_csv(r"C:\Users\HP\Desktop\Pontem Analytics\PyPontem\sample_file.csv")
branch_matrix = pd.read_csv(r"C:\Users\HP\Desktop\Pontem Analytics\PyPontem\test.csv")
```

## Functions in pplparser file
First we have all the functions that we have in tplparser but instead of having average function we have join nodes that allows the user to join the trends of a variable at different branches.

#### The functions that we have and how to write them
metadata function = _extract_metadata

branch profile function = _extract_branch_profiles

catalog function = search_catalog

trends function  = extract_trend

join nodes function = extract_trends_join_nodes

#### Function to extract trends

```python
trends = pplparser.extract_trend(input_matrix)
print(trends)
```

#### Function to join nodes

```python
nodes = pplparser.extract_trends_join_nodes(input_matrix, branch_matrix)
print(nodes)
```

## pplBatchParser

You first load in a list of files to batch parse. Then load it into the instance of pplBatchParser class. 

You can reuse the two csv files or edit them and change to analyze different variables.

### Function in pplBatchParser
trends function = extract_trends

join nodes function = Join_batch_nodes

```python
filepath = [r"C:\Users\HP\Desktop\Pontem Analytics\ppl\PPLfiles_batchparsing\GV-SC_Apr-25-SI_2.ppl",r"C:\Users\HP\Desktop\Pontem Analytics\ppl\PPLfiles_batchparsing\GV-SC_Mar-25-HOC_2.ppl"]
pplbatchparser = pplBatchParser(filepath)
```

#### Function to extract trends of certain variables from a list of files

```python
trends = pplbatchparser.extract_trends(input_matrix)
print(trends)
```

#### Function to join different variable trends from different branches in a list of files

```python
nodes = pplbatchparser.Join_batch_nodes(input_matrix, branch_matrix)
print(nodes)
```
